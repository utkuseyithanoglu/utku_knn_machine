```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python

mm = MinMaxScaler()
x_train_mm = pd.DataFrame(mm.fit_transform(x_train), columns=x_train.columns)
x_test_mm = pd.DataFrame(mm.transform(x_test), columns=x_test.columns)
# i use the minmax scaling before that i did all the outlier 
```

```python
x_train_mm = pd.DataFrame(x_train_mm, columns=x_train.columns)
x_test_mm = pd.DataFrame(x_test_mm, columns=x_test.columns)
#this code is change the column 1234 to normal column name
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train_mm, y_train)

train_preds = knn_model.predict(x_train_mm)
test_preds = knn_model.predict(x_test_mm)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
#this function makes predictions and prints out various evaluation metrics
def model_evaluator(X_matrix, y_target, model):
    preds = model.predict(X_matrix)
    
    print(f"Accuracy score: {round(accuracy_score(y_target, preds),4)}")
    print(f"Precision score: {round(precision_score(y_target, preds, average='macro'),4)}")
    print(f"Recall score: {round(recall_score(y_target, preds, average='macro'),4)}")
    print(f"F1 score: {round(f1_score(y_target, preds, average='macro'),4)}")
    print(f"Confusion matrix: \n{confusion_matrix(y_target, preds)}")
    print(f"Classification report: \n{classification_report(y_target, preds)}")
```

```python
model_evaluator(x_train_mm, y_train, knn_model)
```

```python
model_evaluator(x_test_mm, y_test, knn_model)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyperparameters_dict = {'n_neighbors': [1,3,5,7,15],
                        'weights': ['uniform', 'distance'],
                        'metric': ['euclidean','manhattan','cosine']}
```

```python
gridsearch_model = GridSearchCV(estimator=KNeighborsClassifier(),
                                param_grid=hyperparameters_dict,
                                verbose=3,
                                scoring = 'accuracy')
```

```python
best_model = gridsearch_model.best_estimator_
```
